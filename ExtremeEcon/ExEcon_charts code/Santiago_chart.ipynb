```python
import pandas as pd
import altair as alt
```

```python
c_prod = pd.read_csv('/Users/sambickel-barlow/Downloads/copper-production.csv')
c_uv = pd.read_csv('/Users/sambickel-barlow/Downloads/copper-unit-value.csv')
ann_gdp_2015 = pd.read_csv('/Users/sambickel-barlow/Downloads/API_NY.GDP.MKTP.KD_DS2_en_csv_v2_37980/API_NY.GDP.MKTP.KD_DS2_en_csv_v2_37980.csv', skiprows=4)
cpi_2015 = 237.017
cpi_1998 = 163
```

```python
ann_gdp_2015
```

```python
c_merge = c_prod.merge(c_uv, on=['Year'],how='left')
```

```python
c_merge['c_value'] = c_merge['production|Copper|Mine|tonnes'] * c_merge['unit value|Copper|Mine|constant 1998 US$ per tonne']
```

```python
c_merge = c_merge[~c_merge['c_value'].isna()]
```

```python
c_merge['c_value_2015'] = c_merge['c_value'] * (cpi_2015 / cpi_1998)
```

```python
ann_gdp_2015 = ann_gdp_2015.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], var_name='Year', value_name='GDP_2015')
```

```python
c_merge['Year'] = c_merge['Year'].astype(str)
```

```python
c_merge_gdp = c_merge.merge(ann_gdp_2015[['Country Name', 'Year', 'GDP_2015']], how='left', left_on=['Entity_x','Year'], right_on=['Country Name','Year'])
```

```python
c_merge_gdp['c_value_gdp'] = c_merge_gdp['c_value_2015'] / c_merge_gdp['GDP_2015']
```

```python
c_merge_gdp = c_merge_gdp[c_merge_gdp['Entity_x'] != 'World']
```

```python
chil_ex_23 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1.csv')
chil_ex_18 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1 2.csv')
chil_ex_19 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1 3.csv')
chil_ex_20 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1 4.csv')
chil_ex_21 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1 5.csv')
chil_ex_22 = pd.read_csv('/Users/sambickel-barlow/Downloads/product-exports-1 6.csv')
```

```python
chil_ex = chil_ex_18.merge(chil_ex_19, how='outer', on=['HS4','HS4 ID'])
chil_ex.rename(columns={'Current Exports Value_x': '2018'}, inplace=True)
chil_ex.rename(columns={'Current Exports Value_y': '2019'}, inplace=True)
chil_ex = chil_ex.merge(chil_ex_20, how='outer', on=['HS4','HS4 ID'])
chil_ex.rename(columns={'Current Exports Value': '2020'}, inplace=True)
chil_ex = chil_ex.merge(chil_ex_21, how='outer', on=['HS4','HS4 ID'])
chil_ex.rename(columns={'Current Exports Value': '2021'}, inplace=True)
chil_ex = chil_ex.merge(chil_ex_22, how='outer', on=['HS4','HS4 ID'])
chil_ex.rename(columns={'Current Exports Value': '2022'}, inplace=True)
chil_ex = chil_ex.merge(chil_ex_23, how='outer', on=['HS4','HS4 ID'])
chil_ex.rename(columns={'Current Exports Value': '2023'}, inplace=True)
```

```python
chil_ex = chil_ex.melt(id_vars=['HS4', 'HS4 ID'], var_name='Year', value_name='Export Value')
```

```python
# Consolidate HS4s outside the top 49 in 2023 into 'Other' for ex_chart
top_49_hs4 = (chil_ex[chil_ex['Year'] == '2023']
                .sort_values('Export Value', ascending=False)
                .head(832)['HS4'])
chil_ex_grouped = chil_ex.copy()
chil_ex_grouped['HS4_grouped'] = chil_ex_grouped['HS4'].where(chil_ex_grouped['HS4'].isin(top_49_hs4), 'Other')
```

```python
# Aggregate Export Value by HS4_grouped and Year
chil_ex_grouped_sum = chil_ex_grouped.groupby(['HS4_grouped', 'Year'], as_index=False)['Export Value'].sum()
```

```python
# Calculate percent of total export value for each HS4_grouped in each year
chil_ex_grouped_sum['Year_total'] = chil_ex_grouped_sum.groupby('Year')['Export Value'].transform('sum')
chil_ex_grouped_sum['Export Percent'] = chil_ex_grouped_sum['Export Value'] / chil_ex_grouped_sum['Year_total']
```

```python
# Define the top 4 copper products explicitly
top_copper_products = ['Copper Ore', 'Raw Copper', 'Refined Copper', 'Copper Wire']
# Find all HS4s containing 'Copper'
copper_hs4s = chil_ex['HS4'][chil_ex['HS4'].str.contains('Copper')].unique().tolist()
# Assign 'Other Copper Products' to copper HS4s not in the top 4
def copper_legend_label(x):
    if x in top_copper_products:
        return x
    elif x in copper_hs4s:
        return 'Other Copper Products'
    else:
        return 'All Other exports'
# Update color mapping: top 4 copper products get orange shades, 'Other Copper Products' gets a distinct orange, all others grey
orange_palette = ['#FFA500', '#FF8C00', '#FFD580', '#DA6C06']
grey_color = '#888888'
other_copper_color = '#FFB347'
product_color_map = {
    top_copper_products[0]: orange_palette[0],
    top_copper_products[1]: orange_palette[1],
    top_copper_products[2]: orange_palette[2],
    top_copper_products[3]: orange_palette[3],
    'Other Copper Products': other_copper_color,
    'All Other exports': grey_color
}
# Create a legend group: assign labels using the function above
chil_ex_grouped_sum['HS4_legend'] = chil_ex_grouped_sum['HS4_grouped'].apply(copper_legend_label)
# Ensure Export Percent is float for Altair and scale to 0-100 for percent axis
chil_ex_grouped_sum['Export Percent'] = chil_ex_grouped_sum['Export Percent'].astype(float)
# Chart for all other lines (no legend)
other_chart = alt.Chart(chil_ex_grouped_sum).mark_line(interpolate='monotone').encode(
    x=alt.X('Year', scale=alt.Scale(padding=0), axis=alt.Axis(title=None, labelAngle=0, domainColor='dimgrey', labelColor='#00a767', tickColor='whitesmoke')),
    y=alt.Y('Export Percent:Q', title='% of Total Export Value', axis=alt.Axis(titleAngle=0, titleAnchor='start', titleY=-10, titleX=-50, titleColor='whitesmoke', format='.0%', grid=False, domainColor='dimgrey', labelColor='#00a767', tickColor='whitesmoke')),
    color=alt.value(grey_color),
    detail='HS4_grouped:N',
    tooltip=['HS4_grouped', 'Year', alt.Tooltip('Export Percent:Q', format='.0%')]
 ).transform_filter(
    (alt.datum.HS4_legend == 'All Other exports') & (alt.datum.HS4_grouped != 'All Other exports')
)
# Chart for copper products and 'All Other exports' (for legend)
main_chart = alt.Chart(chil_ex_grouped_sum).mark_line(interpolate='monotone').encode(
    x=alt.X('Year', scale=alt.Scale(padding=0)),
    y=alt.Y('Export Percent:Q'),
    color=alt.Color('HS4_legend', scale=alt.Scale(domain=list(product_color_map.keys()), range=list(product_color_map.values())), legend=alt.Legend(title='Product')),
    tooltip=[alt.Tooltip('HS4_grouped', title='Export'), 'Year', alt.Tooltip('Export Percent:Q', format='.2%')]
 ).transform_filter(
    (alt.datum.HS4_legend != 'All Other exports') | (alt.datum.HS4_grouped == 'All Other exports')
)
# Custom Altair theme for dark background and colored axes/legend
custom_config = alt.Config(
    background='#122b39',
    view=alt.ViewConfig(stroke='transparent'),
    axis=alt.AxisConfig(
        labelColor='#00a767',
        domainColor='dimgrey',
        tickColor='whitesmoke',
        titleColor='whitesmoke'
    ),
    legend=alt.LegendConfig(
        labelColor='#00a767',
        titleColor='#179fdb'
    ),
    title=alt.TitleConfig(
        anchor='start',
        color='#179fdb',
        fontSize=22,
        fontWeight='bold'
    )
)
# Overlay both charts (copper on top), apply config, and add title
ex_chart = (other_chart + main_chart).configure(**custom_config.to_dict()).properties(title='Chilean exports', width=720, height=300)
ex_chart
```
