```python
aceh_gdp = pd.read_csv('data/Aceh/acehGDP.csv')
```

```python
# Base chart without configuration
base = alt.Chart(aceh_gdp).encode(
    x=alt.X('date:N',
            title=None,
            axis=alt.Axis(
                labelAngle=0,
                domainColor="dimgrey",
                labelColor="#333333",
                tickColor="#333333",
                labelFontWeight="bold",
                labelFontSize=13
            ))
)

# Bar layer
bars = base.mark_bar(
    color="#002fa7",
    strokeWidth=2,
    opacity=1
).encode(
    y=alt.Y('growth:Q',
            title=None,
            axis=alt.Axis(
                format=".1%",
                grid=False,
                domainColor="dimgrey",
                labelColor="#333333",
                tickColor="#333333",
                labelFontWeight="bold",
                labelFontSize=12
            ))
)

# Text layer
text = base.mark_text(
    align="center",
    dy=-8,
    color="#333333",
    font="Newcastle Basic Clean, Arial, sans-serif",
    fontWeight="bold",
    fontSize=12
).encode(
    y=alt.Y('growth:Q'),
    text=alt.Text('label:N')
)

# Combine layers and apply configuration to the final chart
final_chart = (bars + text).properties(
    title=alt.TitleParams(
        text="Aceh - growth rates",
        fontSize=15,
        subtitle=["Annual average, 4 year periods, non-oil GDP", ""],
        subtitleFontStyle="italic",
        subtitleFontSize=13,
        anchor="start",
        color="#333333",
        subtitleColor="#333333",
        font="Newcastle Basic Clean, Arial, sans-serif"
    ),
    width=500,
    height=350
).resolve_scale(
    x='shared', y='shared'
).configure(
    background="#fbfbfb",
    view=alt.ViewConfig(stroke="transparent"),
    font="Newcastle Basic Clean, Arial, sans-serif"
)

final_chart
```

```python
# Save to png
final_chart.save('Aceh.png', scale_factor=2)
final_chart.save('Aceh.json', scale_factor=2)
```
