```python
zaatari_data = pd.DataFrame(
    {'Year': [2019, 2020, 2021, 2022, 2023, 2024],
     'Work Permits': [6695, 7676, 3877, 2297, 2961, 1071]}
)
```

```python
# Create the Zaatari work permits chart
base = alt.Chart(zaatari_data).properties(
    width=300,
    height=300,
    title=alt.TitleParams(
        text="Zaatari refugee camp â€“ work permits",
        subtitle=["Work permits issued per year", ""],
        fontSize=15,
        subtitleFontStyle="italic",
        subtitleFontSize=12,
        anchor="start",
        color="#333333",
        subtitleColor="#333333",
        font="Newcastle Basic Clean, Arial, sans-serif"
    )
)

# Create the line layer with formatted tooltips
line = base.mark_line(
    color="#002fa7",
    strokeWidth=2,
    opacity=1,
    point=alt.OverlayMarkDef(
        filled=True,
        fill="#002fa7",
        size=60,
        strokeWidth=2,
        stroke="white",
        opacity=1
    )
).encode(
    x=alt.X('Year:N',
            title=None,
            sort=['2019', '2020', '2021', '2022', '2023', '2024'],
            axis=alt.Axis(
                labelAngle=0,
                domainColor="dimgrey",
                labelColor="#333333",
                tickColor="#333333",
                labelFontWeight="bold",
                labelFontSize=13
            )),
    y=alt.Y('Work Permits:Q',
            title=None,
            axis=alt.Axis(
                format="~s",
                grid=False,
                domainColor="dimgrey",
                labelColor="#333333",
                tickColor="#333333",
                labelFontWeight="bold",
                labelFontSize=13
            )),
    tooltip=[
        alt.Tooltip('Year:N', title='Year'),
        alt.Tooltip('Work Permits:Q', title='Work Permits', format=',.0f')
    ]
)

# Combine layers and configure
final_chart = (line).configure(
    background="#fbfbfb",
    view={"stroke": "transparent"},
    font="Newcastle Basic Clean, Arial, sans-serif"
)

final_chart
```

```python
# Save to png
final_chart.save('Zaatari.png', scale_factor=2)
final_chart.save('Zaatari.json', scale_factor=2)
```
