```python
import pandas as pd
import altair as alt
```

```python
# Load data from a local CSV file
file_path = 'Portfolio/Legislature Age guess.csv'
L_ages = pd.read_csv(file_path)
```

```python

label_list = ['United States' , 'Mexico' , 'Australia' , 'Ireland' , 'South Korea' , 'Japan' , 'U.K.' , 'Greece' , 'Canada' , 'Norway' , 'Switzerland' , 'Germany' , 'Italy' , 'Netherlands' , 'Sweden']

L_ages['Country_labels'] = L_ages.apply(
    lambda row: row['Country'] if row['Country'] in label_list else '', axis=1
)

L_ages
```

```python
# Create a Vega-Lite chart with Altair using the loaded data
L_ages_chart = (
alt.Chart(L_ages[L_ages['Country'] == 'United States']).mark_circle(color='red') #Create the scatter chart and add the red point for the US data point
.encode(
    x={
        "field":"Median Age of Population", #Specify "Median Age of Population" for the x-axis
        "type":"quantitative", #Specify a quantitative data type
        "scale":alt.Scale(domain=[30, 60]), #Set the x-axis scale to between 30 and 60
        "axis":alt.Axis( #Set x-axis parameters
            gridDash=[5, 5], #Put lines between each 5 ages on x-axis
            title="Median Age of Population", #Change x-axis title
            titleFont='Georgia', #Change x-axis title font
            titleFontSize=10, #Resize x-axis title
            labelFontWeight='lighter', #Make x-axis title less bold
        )
    },
    y={
        "field":"Average Age of Lawmakers", #Specify "Average Age of Lawmakers" for y-axis
        "type":"quantitative", #Specify a quantitative data type
        "scale":alt.Scale(domain=[30, 60]), #Set the y-axis scale to between 30 and 60
        "axis":alt.Axis( #Set y-axis parameters
            gridDash=[5, 5], #Put lines between each 5 ages on y-axis
            title="Average Age of Lawmakers", #Change y-axis title
            titleFont='Georgia', #Change x-axis title font
            titleFontSize=10, #Resize x-axis title
            labelFontWeight='lighter', #Make x-axis title less bold
        )
    }
) 
+ 
alt.Chart(L_ages[L_ages['Country'] != 'United States']).mark_circle(color='black') #Adds points for all non-US data points and colors them black
.encode(
    x={
        "field":"Median Age of Population", #Specify "Median Age of Population" for the x-axis
        "type":"quantitative", #Specify a quantitative data type
        "scale":alt.Scale(domain=[30, 60]), #Set the x-axis scale to between 30 and 60
    },
    y={
        "field":"Average Age of Lawmakers", #Specify "Average Age of Lawmakers" for y-axis
        "type":"quantitative", #Specify a quantitative data type
        "scale":alt.Scale(domain=[30, 60]), #Set the y-axis scale to between 30 and 60
    }
) 
+ 
alt.Chart(L_ages[L_ages['Country'] == 'United States']).mark_text(align='left', dx=5, dy=-5, color='red') #Add red label US data point 
.encode(
    x="Median Age of Population",
    y="Average Age of Lawmakers",
    text="Country_labels",
    )
+ 
alt.Chart(L_ages[L_ages['Country'] != 'United States']).mark_text(align='left', dx=5, dy=-5, color='black') #Add black label for non-US data points that were specified in the original chart (some data points have no label)
.encode(
    x="Median Age of Population",
    y="Average Age of Lawmakers",
    text="Country_labels"
    )
+
alt.Chart(pd.DataFrame({'x': [30, 60]})).mark_line(color='black' , strokeWidth = 1) #Add a 45 degree line to show where Lawmaker age is equal to median pop age
.encode(
    x='x',
    y='x'
)
+
alt.Chart(pd.DataFrame({ #Add a textbox label for the 45 degree line
    'x': [50],
    'y': [61],
    'text': ['If age of lawmakers and politicans were similar']
})).mark_text(
    align='left',
    baseline='middle',
    fontSize=8,
    font='Georgia'
).encode(
    x='x:Q',
    y='y:Q',
    text='text:N'
)
)

L_ages_chart = L_ages_chart.configure_title( #Add a title to the chart
    fontSize=15,
    font='Georgia',
    anchor='start',
    color='black'
).properties(
    title="One Good Chart: America's Political Gerontocracy"
)
L_ages_chart
```
